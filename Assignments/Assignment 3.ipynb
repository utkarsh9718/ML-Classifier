__6. Formatting Data – making data types compaitable with other data types__

```python
dataset.diagnosis.unique()
```

```python
def modify_diagnosis(series):
    modify=[]
    for  i in series.values:
        if i=='B':
            modify.append(0)
        elif i=='M':
            modify.append(1)
        else:
            modify.append(np.nan)
    return pd.Series(modify)
```

```python
diag=modify_diagnosis(dataset.diagnosis)
```

```python
dataset.diagnosis=diag
```
